```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
oil_data = pd.read_csv("oil2.csv")
print(oil_data.head(5))
```

```python
oil_data.describe()
```

```python
sns.pairplot(oil_data)
```

```python
#get columns which give features of the data
oil_data.columns
```

```python
#make x, y arrays necessary for splitting 
x = oil_data[['PMS','AGO','Kero','Price Per Barell','Int Price in KSH']]
y = oil_data['Local Price in KSH']
```

```python
from sklearn.model_selection import train_test_split
#splitting the data for training
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
#import linear regression algorithm
from sklearn.linear_model import LinearRegression
#create an instance object for linear regression
model = LinearRegression()
#fit the data
model.fit(x_train,y_train)
```

```python
#examine the coefficients
pd.DataFrame(model.coef_,x.columns,columns=['Coeff'])
```

```python
#examine the intercepts
print(model.intercept_)
```

```python
#make your predictions
predictions = model.predict(x_test)
print(predictions)
```

```python
#assess the score of your model
model.score(x_test,y_test)
```

```python
#compare x_test values and y_values using a scatter plot 
plt.scatter(y_test,predictions)
```

```python
#plot the residuals to asses the performance of our model
plt.hist(y_test - predictions)
```

```python
#testing the performance of our model
from sklearn import metrics
metrics.mean_absolute_error(y_test,predictions)
```
